```python
import numpy as np

plt.figure(1)
data_intel = ldata[ldata['Cpu_Company'] == 'Intel']['Company'].value_counts()
data_amd = ldata[ldata['Cpu_Company'] == 'AMD']['Company'].value_counts()
n_ticks_i = np.arange(len(data_intel))
n_ticks_a = np.arange(len(data_amd))
offset = 0.2
w = 0.4

plt.bar(n_ticks_i-offset, data_intel, width=w)
plt.bar(n_ticks_a+offset, data_amd, width=w)
plt.title('Соотношение производителей')
plt.xlabel('Компании')
plt.ylabel('Кол-во')
plt.xticks(n_ticks_i, data_intel.index,rotation=50)
plt.grid()
plt.legend(['Intel', 'AMD'])
```

```python
plt.figure(1)
data_intel = ldata[ldata['Cpu_Company'] == 'Intel']['Company'].value_counts()
data_intel = data_intel.apply(lambda d: d/data_intel.sum()*100)
data_amd = ldata[ldata['Cpu_Company'] == 'AMD']['Company'].value_counts()
data_amd = data_amd.apply(lambda d: d/data_amd.sum()*100)
n_ticks_i = np.arange(len(data_intel))
n_ticks_a = np.arange(len(data_amd))
offset = 0.2
w = 0.4

plt.bar(n_ticks_i-offset, data_intel, width=w)
plt.bar(n_ticks_a+offset, data_amd, width=w)
plt.title('Соотношение производителей')
plt.xlabel('Компании')
plt.ylabel('Доля, %')
plt.xticks(n_ticks_i, data_intel.index,rotation=50)
plt.grid()
plt.legend(['Intel', 'AMD'])
```

```python
plt.figure(1)
fig, ax = plt.subplots()
cpc = ldata['Cpu_Company'].unique()
data = [ldata[ldata['Cpu_Company'] == c]['Price_euros'] for c in cpc]
ax.boxplot(data)

plt.title('Price_euros vs Cpu_Company')
plt.xlabel('Cpu_Company')
plt.ylabel('Price_euros')
plt.xticks(range(1,len(cpc)+1), cpc)
plt.grid()

plt.figure(2)
fig, ax = plt.subplots()
mt = ldata['Memory_Type'].unique()
data = [ldata[ldata['Memory_Type'] == m]['Price_euros'] for m in mt]
ax.boxplot(data)

plt.title('Price_euros vs Memory_Type')
plt.xlabel('Memory_Type')
plt.ylabel('Price_euros')
plt.xticks(range(1,len(mt)+1), mt)
plt.grid()

plt.figure(3)
fig, ax = plt.subplots()
ram = ldata['Ram'].unique()
data = [ldata[ldata['Ram'] == m]['Price_euros'] for m in ram]
ax.boxplot(data)

plt.title('Price_euros vs RAM')
plt.xlabel('RAM')
plt.ylabel('Price_euros')
plt.xticks(range(1,len(ram)+1), ram)
plt.grid()

plt.figure(4)
fig, ax = plt.subplots()
cmp = ldata['Company'].unique()
data = [ldata[ldata['Company'] == c]['Price_euros'] for c in cmp]
ax.boxplot(data)

plt.title('Price_euros vs Company')
plt.xlabel('Company')
plt.ylabel('Price_euros')
plt.xticks(range(1,len(cmp)+1), cmp, rotation=50)
plt.grid()
```
