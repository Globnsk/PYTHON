```python
Client_choice = df[(df['Company'] == 'MSI') & (df['Gpu'] == 'Nvidia GeForce GTX 1050 Ti') & (df['Inches'] != 15.6)]

print(Client_choice ['Price_euros'].min(), Client_choice['Price_euros'].max())
```
